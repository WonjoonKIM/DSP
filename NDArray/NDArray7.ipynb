```python
#NumPy에서는 물론 우리가 아는 sin, cos, tan 등 trigonometric function들을 한 번에 구할 수 있는 method들도 존재한다. trigonometric function에 대해서 살펴볼 method들은 다음과 같다.
#1. np.sin()
#2. np.cos()
#3. np.tan()
#4. np.arcsin()
#5. np.arccos()
#6. np.arctan()
#7. np.deg2rad()
#8. np.rad2deg()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

PI = math.pi

radian_range = np.linspace(0, 4*PI, 300)

sin_np = np.sin(radian_range)
cos_np = np.cos(radian_range)
tan_np = np.tan(radian_range)

plt.figure(figsize = (8,5))
plt.plot(sin_np, label = 'sin')
plt.plot(cos_np, label = 'cos')
plt.plot(tan_np, label = 'tan')
plt.legend(facecolor = (0.8, 0.8, 0.8))
plt.ylim([-2, 2])
plt.grid()
plt.show()
#위와 같이 우리가 알던 sin, cos, tan 함수를 만들어낼 수 있다.
#NumPy에서 이렇게 함수를 만들 때 좋은 점은, 300개의 point에 대해서 한 번에 sin, cos, tan 값을 구할 수 있다는 점이다.
```

```python
radian_range = np.linspace(-1, 1, 300)

arcsin_np = np.arcsin(radian_range)
arccos_np = np.arccos(radian_range)
arctan_np = np.arctan(radian_range)

plt.figure(figsize = (8,5))
plt.plot(arcsin_np, label = 'arcsin')
plt.plot(arccos_np, label = 'arccos')
plt.plot(arctan_np, label = 'arctan')
plt.legend(facecolor = (0.8, 0.8, 0.8))
plt.ylim([-2, 5])
plt.grid()
plt.show()
```

```python
# degree와 radian을 많이 사용하게 되는데 이때, 이 변환을 해주는 method들이 deg2rad(), rad2deg()이다.
# 이를 확인하기 위해 deg를 직접 rad으로 바꿔보고 deg2rad()을 이용하여 두 값을 비교해보자.
deg_np = np.array([0, 30, 45, 60, 90])
rad_np = deg_np/180*PI

print(rad_np)
print(np.deg2rad(deg_np)) #일치하면옳게 변환된 것

rad_range = np.linspace(0, 4*PI, 300)
sin_np1 = np.sin(rad_range)

deg_range = np.linspace(0, 4*180, 300)
sin_np2 = np.sin(np.deg2rad(deg_range))

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15,5))
ax1.plot(sin_np1)
ax2.plot(sin_np2)
```

```python
#Trigonometric과 함께 많이 사용되는 함수들이 Hyperbolic function들이다. 이는 NumPy에서 다음과 같이 제공하고 있다.
#1. np.sinh()
#2. np.cosh()
#3. np.tanh()
#4. np.arcsinh()
#5. np.arccosh()
#6. np.arctanh()

import numpy as np
import matplotlib.pyplot as plt

radian_range = np.linspace(-3, 3, 300)

sinh_np = np.sinh(radian_range)
cosh_np = np.cosh(radian_range)
tanh_np = np.tanh(radian_range)

plt.figure(figsize = (8,5))
plt.plot(sinh_np, label = 'sinh')
plt.plot(cosh_np, label = 'cosh')
plt.plot(tanh_np, label = 'tanh')
plt.legend(facecolor = (0.8, 0.8, 0.8))
plt.grid()
plt.show()
```

```python
radian_range = np.linspace(-3, 3, 300)

arcsinh_np = np.arcsinh(radian_range)
arccosh_np = np.arccosh(radian_range)
arctanh_np = np.arctanh(radian_range)

plt.figure(figsize = (8,5))
plt.plot(arcsinh_np, label = 'arcsinh')
plt.plot(arccosh_np, label = 'arccosh')
plt.plot(arctanh_np, label = 'arctanh')
plt.legend(facecolor = (0.8, 0.8, 0.8))
plt.grid()
plt.show()
#위에서 arccos은 함수가 아니기 때문에 오류가 뜨는 모습을 볼 수 있고, 다른 arc-hyperbolic function들은 모두 제대로 나오는 것을 확인할 수 있다.
```

```python
# Exponent를 구하는 방법과 그 역연산인 Logarithm에 대한 설명이다.
# 이때 base가 natural constant와 2일때를 다룬다. 이는 우리가 앞으로 사용하는 exp, log는 보통 base가 natural constant일 때를 말하고, 
#우리가 computer science, imformation theory 등을 다룰 때 base는 보통 2이기 때문이다.
# 사실 natural logarithm을 다루는 method들 외에는 모두 np.power()로 대체할 수 있지만 다른 사람이 쓴 코드를 알아보기 위해서라도 간단히 알아보도록 하자.

#1. np.exp()
#2. np.exp2()
#3. np.log()
#4. np.log2()
#5. np.log10()

import numpy as np
import matplotlib.pyplot as plt

#먼저 exponent들을 다뤄보도록 하자. 밑이 natural constant일 때는, derivative나 integral을 구하기가 훨씬 수월하므로 이론적인 분석에서 자주 사용된다.
#그리고 bit를 다룰 때 base가 2를 자주 사용하므로 둘 다 익혀두는 것이 좋다.

x_range = np.linspace(-2, 2, 500)

exp_np = np.exp(x_range)
exp2_np = np.exp2(x_range) #x_range 범위에 대해 500point의 밑을 2로 하는 제곱 수

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(x_range, exp_np, label = 'exp')
ax.plot(x_range, exp2_np, label = 'exp2')
ax.grid()
fig.legend(facecolor = (0.8, 0.8, 0.8), loc = 'upper left')
```

```python
radian_range = np.linspace(-3, 3, 300)

sinh_np1 = (np.exp(radian_range) - np.exp(-1*radian_range))/2 # /2도 마찬가지로 element마다 /2를 해준 결과이다, 
sinh_np2 = np.sinh(radian_range) + 0.5 #일부러 곂치는 것을 방지하기 위해 np.sinh()의 결과에 0.5를 더해줬다

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(radian_range, sinh_np1, label = 'sinh_np1')
ax.plot(radian_range, sinh_np2, label = 'sinh_np2')
ax.grid()
fig.legend(facecolor = (0.8, 0.8, 0.8), loc = 'upper left')
```

```python
#다음은 base가 각각 e, 2, 10인 logarithm들을 구해보도록 하자.

radian_range = np.linspace(0.01, 10, 300)

log_np = np.log(radian_range)
log2_np = np.log2(radian_range)
log10_np = np.log10(radian_range)

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(radian_range, log_np, label = 'log')
ax.plot(radian_range, log2_np, label = 'log2')
ax.plot(radian_range, log10_np, label = 'log10')
ax.grid()
fig.legend(facecolor = (0.8, 0.8, 0.8), loc = 'upper left')
```

```python
#base가 3, 4, 5인 log function들에 대해서 다음과 같이 구현이 가능하다.

radian_range = np.linspace(0.01, 10, 300)

log3_np = np.log(radian_range) / np.log(3)
log4_np = np.log(radian_range) / np.log(4)
log5_np = np.log(radian_range) / np.log(5)

fig, ax = plt.subplots(figsize = (8,4))
ax.plot(radian_range, log3_np, label = '$log_{3}x$')
ax.plot(radian_range, log4_np, label = '$log_{4}x$')
ax.plot(radian_range, log5_np, label = '$log_{5}x$')
ax.grid()
fig.legend(facecolor = (0.8, 0.8, 0.8), loc = 'upper left', fontsize = 'large')
```
